```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('./data_9_5day.csv') #Original 
from sklearn.utils import shuffle
df = shuffle(df).reset_index(drop=True)
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
XU_train,XL,yu_train,yL=train_test_split(X,y,random_state=0,test_size=0.05)
```

```python
len(XU_train)
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler 
```

```python
XL['ProtocolName']=yL
```

```python
a=XL.groupby('ProtocolName')['ProtocolName'].count().sort_values(ascending=False)
```

```python
threshold=10
b=a[a<=threshold]
b
```

```python
XU_train=XU_train.append(XL[XL['ProtocolName'].isin(b.index)].iloc[:,:-1],sort=False)
XU_train=shuffle(XU_train).reset_index(drop=True)
```

```python
len(XU_train)
```

```python
XL
```

```python
len(XL)
```

```python
XL.drop(XL[XL['ProtocolName'].isin(b.index)].index,inplace=True)
```

```python
len(XL)
```

```python
a=XL.groupby('ProtocolName')['ProtocolName'].count().sort_values(ascending=False)
a
```

```python
len(XU_train)+len(XL)
```

```python
b=a.where(a>5000)
b=b.fillna(5000,downcast='infer')
threshold=dict(zip(b.index,b.values))
ros=RandomOverSampler(sampling_strategy=threshold,random_state=0)
Xi=XL.iloc[:,:-1]
yi=XL.iloc[:,-1]
X1,y1=ros.fit_resample(Xi,yi)
```

```python
b=b.where(b<6000)
b=b.fillna(6000,downcast='infer')
threshold=dict(zip(b.index,b.values))
rus=RandomUnderSampler(sampling_strategy=threshold,random_state=0)
```

```python
XL_train,yL_train=rus.fit_resample(X1,y1)
```

```python
ydf=pd.DataFrame(yL_train)
ygrp=ydf.groupby('ProtocolName')['ProtocolName'].count().sort_values(ascending=False)
ygrpdf=pd.DataFrame(ygrp)
ygrpdf['percent']=100*ygrp/sum(ygrp)
ygrpdf
```

```python
XL=XL_train
```

```python
XL['ProtocolName']=yL_train
```

```python
XL.to_csv('9_510_L_95_5.csv',index=False)
XU_train.to_csv('9_510_U_95_5.csv',index=False)
```
